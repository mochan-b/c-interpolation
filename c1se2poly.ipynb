```python
from SE2 import SE2
import math


p1 = SE2 (1, 1, -45)
p2 = SE2 (3, 3, 60)
p3 = SE2 (5, 2, -75)
p4 = SE2 (6, 4, 0)
```

```python
import numpy as np

import sympy
from sympy.abc import a, b, c, d, e, f

def c1quadratic(pt1, pt2):
    t1 = math.tan(math.radians(pt1.theta))
    t2 = math.tan(math.radians(pt2.theta))
    s = sympy.solve([c - pt1.x, f-pt1.y, a + b + c - pt2.x, c + d + e - pt2.y, e/b - t1, (2*d + e) / (2*a + b) - t2])
    t = np.arange(0, 1, 0.01)
    px = lambda t : s[a]*t**2 + s[b]*t + s[c]
    py = lambda t : s[d]*t**2 + s[e]*t + s[f]
    x = list(map(px, t))
    y = list(map(py, t))
    ppx = lambda t : 2*s[a]*t + s[b]
    ppy = lambda t : 2*s[d]*t + s[e]
    dx = list(map(ppx, t))
    dy = list(map(ppy, t))
    return x,y, dx, dy

import matplotlib.pyplot as plt
fig=plt.figure(figsize=(12, 10))

p1.draw(plt)
p2.draw(plt)
p3.draw(plt)
p4.draw(plt)

x1,y1,dx1,dy1 = c1quadratic(p1, p2)
plt.plot(x1,y1)
x2,y2,dx2,dy2 = c1quadratic(p2, p3)
plt.plot(x2, y2)
x3,y3,dx3,dy3 = c1quadratic(p2, p4)
plt.plot(x3,y3)
```

```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(12, 10))

t = np.arange(0, 1, 0.01)
x1,y1,dx1, dy1 = c1quadratic(p1, p2)
d1 = list(map(math.atan, np.divide(dy1, dx1)))
plt.plot(t*(p2.x - p1.x)+p1.x,d1)
x2,y2,dx2,dy2 = c1quadratic(p2, p3)
d2 = list(map(math.atan, np.divide(dy2, dx2)))
plt.plot(t*(p3.x - p2.x)+p2.x, d2)
x3,y3,dx3,dy3 = c1quadratic(p2, p4)
d3 = list(map(math.atan, np.divide(dy3, dx3)))
plt.plot(t*(p4.x - p2.x)+p2.x,d3)
```
