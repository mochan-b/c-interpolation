```python
# This is the point we are trying to draw right now
x = 1
y = 2
theta = 45
kappa = 0.6

# Draw the point and the circle of curvature
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import math

fig=plt.figure(figsize=(10, 8))

plt.scatter(x, y, s=100, color='black')
plt.arrow(x, y, 0.25 * math.cos(math.radians(theta)), 0.25 * math.sin(math.radians(theta)), color='red', linewidth=2)

# Calculate all the relevant values for drawing the curvature
radius = 1.0/math.fabs(kappa)
normal_angle = 90 * math.copysign(1, kappa)
cx = x + radius/2 * math.cos(math.radians(theta+normal_angle))
cy = y + radius/2 * math.sin(math.radians(theta+normal_angle))

angle = theta - normal_angle
if math.copysign(1, kappa) == -1 :
    theta1 = -20
    theta2 = 0
else:
    theta1 = 0
    theta2 = 20

ax = plt.gca()
arc = mpatches.Arc((cx,cy), radius, radius, angle = angle, theta1 = theta1, theta2 = theta2, color='blue', linestyle='--', linewidth=3)
ax.add_patch(arc)

plt.axis('scaled')

# Draw the circle as well of radius 1
```
